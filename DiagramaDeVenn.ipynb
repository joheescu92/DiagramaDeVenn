```python
#Importamos y cargamos el dataframe
from matplotlib_venn import venn2, venn2_unweighted
from matplotlib_venn import venn3, venn3_unweighted
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
from google.colab import drive
drive.mount('/content/drive')
df =pd.read_csv('/content/drive/MyDrive/data/animales_caracteristicas.csv')
df
```

```python
# ahora vamos a armar  conjuntos y se van a llamar de la siguiente manera:
#animales que respiran
#animales que tienen piernas
#animales que nadan
# luego filtramos
respira = set(df[df['respira'] == 1]['animal'])
respira
```

```python
# ahora creo el conjunto de los animales que tienen piernas
piernas =set(df[df['piernas']==1]['animal'])
piernas
```

```python
# ahora creo el conjunto de los animales que  nadan
nada =set(df[df['nada']==1]['animal'])
nada
```

```python
venn3_unweighted(
    [
        respira,
        piernas,
        nada
    ],
    set_labels=('Respira Aire ', 'Tiene Piernas','Puede Nadar') # le doy nombres nuevos a esos conjuntos
)
plt.title('Animales') #titulo que le pongo al gr√°fico
```
